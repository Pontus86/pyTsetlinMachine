```python
fpr, tpr, _ = roc_curve(actual, preds)
print("AUC =", metrics.auc(fpr, tpr))
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```

```python
#1 sortera efter lägst värde på prediction.
#2 hitta hur många man får missa givet en viss sensitivitet
#3 hitta den med hjärtinfarkt med lägst score efter de man får missa
#4 sätt max score för rule out till < den i 3
#5 räkna ut hur många totalt och i % som har lägre än denna score.
#6 räkna antalet utan hjärtinfarkt som skickats hem / alla som skickats hem = NPV

total = pd.DataFrame(preds, columns = ["Predictions"])
total["Actual"] = actual.tolist()
total = total.sort_values(by=['Predictions'])
total.head()

sensitivity = 0.99
print(total["Actual"].sum())
print(len(total))
allowed_miss = np.floor(total["Actual"].sum() * (1-sensitivity)).astype(int)
print(allowed_miss)
#total[total.Actual == 1].head(100)

#Hitta prediktionsvärdet på den som var första över gränsen
lowest_prediction = total[total.Actual == 1].iloc[allowed_miss].Predictions
print("Rule out threshold =", lowest_prediction)
#total[total.Actual == 1].index[0]
ruled_out = len(total[total.Predictions < lowest_prediction])
print("Ruled out = ", ruled_out, "(", np.round(100*ruled_out/len(total), 2), "% )")
print("NPV = ", np.round(100*ruled_out/(ruled_out+allowed_miss), 2), " %")
```

```python
#Validation of the results using the same threshold as the dev-data

data_val = pd.read_csv('../../test-preds.massaged.csv')
data_val.gender[data_val.gender == 'Male'] = 1
data_val.gender[data_val.gender == 'Female'] = 0

X_val_ = data_val.loc[:, 'gender':'bl-Tn1']
X_val_ = X_val_.to_numpy()
Y_val = data_val.loc[:,'label-index-mi'].to_numpy()


b_val = Binarizer(max_bits_per_feature = 100)
b_val.fit(X_val_)
X_val = b_val.transform(X_val_)
pd.DataFrame(X_val_).head()

preds = tm.predict(X_val)
actual = Y_val

total_val = pd.DataFrame(preds, columns = ["Predictions"])
total_val["Actual"] = actual.tolist()

allowed_miss = np.floor(total_val["Actual"].sum() * (1-sensitivity)).astype(int)
print(allowed_miss)
ruled_out = len(total_val[total_val.Predictions < lowest_prediction])
print("Rule out threshold =", lowest_prediction)
print("Ruled out = ", ruled_out, "(", np.round(100*ruled_out/len(total_val), 2), "% )")
print("NPV = ", np.round(100*ruled_out/(ruled_out+allowed_miss), 2), " %")
```

```python
from pyTsetlinMachine.tm import MultiClassTsetlinMachine
from pyTsetlinMachine.tools import Binarizer
import numpy as np

from sklearn import datasets
from sklearn.model_selection import train_test_split

breast_cancer = datasets.load_breast_cancer()
X = breast_cancer.data
Y = breast_cancer.target

b = Binarizer(max_bits_per_feature = 10)
b.fit(X)
X_transformed = b.transform(X)

tm = MultiClassTsetlinMachine(800, 40, 5.0)

print("\nMean accuracy over 100 runs:\n")
tm_results = np.empty(0)
for i in range(100):
	X_train, X_test, Y_train, Y_test = train_test_split(X_transformed, Y, test_size=0.2)

	tm.fit(X_train, Y_train, epochs=25)
	tm_results = np.append(tm_results, np.array(100*(tm.predict(X_test) == Y_test).mean()))
	print("#%d Average Accuracy: %.2f%% +/- %.2f" % (i+1, tm_results.mean(), 1.96*tm_results.std()/np.sqrt(i+1)))
```
